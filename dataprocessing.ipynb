```python
import pandas as pd
```

```python
# 데이터 전처리
# 크롤링해서 저장한 엑셀파일을 불러옵니다.
Tgeni = pd.read_excel("../Tsherpa_reviews.xlsx")
Tprice = pd.read_excel("../Tsheprice.xlsx")
Tdiscount = pd.read_excel("../Tshedprice.xlsx")
```

```python
# column명을 정의합니다.
Tgeni = Tgeni.rename(columns={'별점':'star', '제목':'title','시간':'time'})
Tprice = Tprice.rename(columns={'카테고리':'category', '제목':'title','가격':'price'})
Tdiscount = Tdiscount.rename(columns={'카테고리':'category', '제목':'title','price':'D_price'})
```

```python
## Tprice의 값을 int로 변환합니다.
Tprice['price'] = Tprice['price'].replace('50,000 원','50000')
Tprice['price'] = Tprice['price'].replace('30,000 원','30000')
Tprice['price'] = Tprice['price'].replace('70,000 원','70000')
Tprice['price'] = Tprice['price'].replace('40,000 원','40000')
Tprice['price'] = Tprice['price'].replace('15,000 원','15000')
Tprice['price'] = Tprice['price'].astype(int)
type(Tprice['price'])
```

```python
## Tgeni의 star를 int로 변환
Tgeni['star'] = Tgeni['star'].replace('5점','5')
Tgeni['star'] = Tgeni['star'].replace('4점','4')
Tgeni['star'] = Tgeni['star'].replace('3점','3')
Tgeni['star'] = Tgeni['star'].replace('2점','2')
Tgeni['star'] = Tgeni['star'].replace('1점','1')


Tgeni['star'] = Tgeni['star'].astype(int)
type(Tgeni['star'])
```

```python
## Tgeni의 time을 int로 변환
Tgeni['time'] = Tgeni['time'].replace('30시간','30')
Tgeni['time'] = Tgeni['time'].replace('15시간','15')
Tgeni['time'] = Tgeni['time'].replace('5시간','5')
Tgeni['time'] = Tgeni['time'].replace('20시간','20')
Tgeni['time'] = Tgeni['time'].replace('4시간','4')
Tgeni['time'] = Tgeni['time'].replace('10시간','10')
Tgeni['time'] = Tgeni['time'].replace('6시간','6')
Tgeni['time'] = Tgeni['time'].replace('0시간','0')
Tgeni['time'] = Tgeni['time'].replace('3시간','3')
Tgeni['time'] = Tgeni['time'].replace('16시간','16')
Tgeni['time'] = Tgeni['time'].replace('1시간','1')

Tgeni['time'] = Tgeni['time'].astype(int)
type(Tgeni['time'])
```

```python
## Tprice와 Tsale을 병합합니다.
df1 = pd.DataFrame(Tprice)
df2 = pd.DataFrame(Tdiscount)

price = pd.merge(df1, df2, on=['title', 'category'])
price
```

```python
## Tgeni와 price를 병합합니다.
Tshera = pd.merge(price, Tgeni, on='title')
Tshera
```

```python
# 후기 개수를 저장합니다.
Tshera['review_c'] = Tshera.groupby('title')['title'].transform('size')

# 평점을 저장합니다.
Tshera['star_m'] = Tshera.groupby('title')['star'].transform('sum') / Tshera['review_c']

# 별점을 제거합니다.
Tshera.drop(columns=['star'], inplace=True)
Tshera
```

```python
#중복된 값을 제거하고, 회사변수를 추가합니다.
Tshera.drop_duplicates(inplace=True)
Tshera['company'] = 'T'
Tshera
```

```python
#전처리후 데이터를 엑셀로 저장합니다.
Tshera.to_excel('Tsherpa_data.xlsx', index=False)
```

```python
# 아이스크림 교육원의 데이터를 불러옵니다.
dprice = pd.read_excel("../i-dprices.xlsx")
ice = pd.read_excel("../i-price.xlsx")
```

```python
# column명을 바꿉니다.
dprice = dprice.rename(columns={'강의명':'title','가격':'D_price'})
ice = ice.rename(columns={'Class Name':'title', 'Category':'category', '차시':'time','가격':'price'})
```

```python
# int로 변환
dprice['D_price'] = dprice['D_price'].replace('85,000원','85000')
dprice['D_price'] = dprice['D_price'].replace('50,000원','50000')
dprice['D_price'] = dprice['D_price'].replace('15,000원','15000')
dprice['D_price'] = dprice['D_price'].replace('12,000원','12000')
dprice['D_price'] = dprice['D_price'].replace('140,000원','140000')
dprice['D_price'] = dprice['D_price'].replace('21,000원','21000')
dprice['D_price'] = dprice['D_price'].replace('30,000원','30000')
dprice['D_price'] = dprice['D_price'].replace('75,000원','75000')
dprice['D_price'] = dprice['D_price'].replace('무료','0')
dprice['D_price'] = dprice['D_price'].replace('40,000원','40000')
dprice['D_price'] = dprice['D_price'].replace('130,000원','130000')
dprice['D_price'] = dprice['D_price'].replace('65,000원','65000')
dprice['D_price'] = dprice['D_price'].replace('35,000원','35000')

dprice['D_price'] = dprice['D_price'].astype(int)
type(dprice['D_price'])
```

```python
# int로 변환
ice = ice.dropna()
ice['price'] = ice['price'].replace('85,000원','85000')
ice['price'] = ice['price'].replace('50,000원','50000')
ice['price'] = ice['price'].replace('15,000원','15000')
ice['price'] = ice['price'].replace('12,000원','12000')
ice['price'] = ice['price'].replace('140,000원','140000')
ice['price'] = ice['price'].replace('21,000원','21000')
ice['price'] = ice['price'].replace('30,000원','30000')

ice['price'] = ice['price'].astype(int)
ice['Rating'] = ice['Rating'].astype(int)
ice['time'] = ice['time'].astype(int)
```

```python
#title을 기준으로 데이터를 합칩니다.
df1 = pd.DataFrame(ice)
df2 = pd.DataFrame(dprice)
icream = pd.merge(df1, df2, on='title')

#후기 개수변수를 설정합니다.
icream['review_c'] = icream.groupby('title')['title'].transform('size')

# 평균 별점을 설정합니다.
icream['star_m'] = icream.groupby('title')['Rating'].transform('sum') / icream['review_c']

# 기존 별점을 삭제하고, 겹치는 값들을 삭제합니다.
icream.drop(columns=['Rating'], inplace=True)
icream.drop_duplicates(inplace=True)

icream['company'] = 'I'
icream
```

```python
icream.to_excel('i-cream_data.xlsx', index=False)
```

```python
#데이터 병합
total = pd.concat([icream, Tshera], ignore_index=True)
total
```
